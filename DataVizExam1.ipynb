###f. Create a new visualization that I have not asked for. It can be anything but you should be creative.

```python
ax = sns.barplot(data = df,
                 x = 'minutes',
                 y = 'Goals_Assists')
ax.set_title('2022 World Cup Goals + Assists by Minutes Played')
ax.set_xlabel('Minutes Played')
ax.set_ylabel('Goals + Assists')
plt.show()
```

I wanted to see how effective a player might be so I wanted to see time versus goals and assists and you can see the guy that score a bunch of points played for a really long time, the problem with this graph though is that you cannot read anything. I think the best way to make the graph for this would have been to use a scatterplot but we have already used a scatter plot in the previous question so I chose to see if you could make a nice bar chart, I just assumed it would just group the numbers at the bottom of it by parts in a range starting at 0 and going to whatever the max was but no it went and labeled every single one.

```python
bins = [0, 90, 180, 270, 360, 450, 540]
labels = ['0-90', '91-180', '181-270', '271-360', '361-450', '451-540']
df['minutes_binned'] = pd.cut(df['minutes'], bins=bins, labels=labels, right=False)

# Calculate the average Goals_Assists for each minute bin
binned_data = df.groupby('minutes_binned')['Goals_Assists'].mean().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(data = binned_data,
                 x = 'minutes_binned',
                 y = 'Goals_Assists')
ax.set_title('Average Goals + Assists by Minutes Played Bins')
ax.set_xlabel('Minutes Played Bins')
ax.set_ylabel('Average Goals + Assists')
plt.show()
```
