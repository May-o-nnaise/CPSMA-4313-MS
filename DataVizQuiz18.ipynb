#Quiz 18

```python
import pandas as pa
import datetime as datetime

df = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/IOT-temp.csv")
df.head(10)
```

##Convert the noted_date into DateTime, it is in day-month-Year Hour:Minute format.

```python
pa.to_datetime(df.noted_date, format='%d-%m-%Y %H:%M')
```

##What is the minimum and maximum date and time?

```python
df['noted_date'] = pa.to_datetime(df.noted_date, format='%d-%m-%Y %H:%M')
print(f"Minimum Date: {df['noted_date'].min()}")
print(f"Maximum Date: {df['noted_date'].max()}")
```

take the dable and grab print the minimum and maximum

##What is the largest interval inbetween temperature checks?

```python
df['noted_date'] = pa.to_datetime(df.noted_date, format='%d-%m-%Y %H:%M')
largest_interval = df['noted_date'].sort_values().diff().max()
print(f"Largest Interval between temperature checks: {largest_interval}")
```

this sorts the values by difference and takes the maximum difference.

##What is the average datetime?

```python
avg_datetime = df['noted_date'].mean()
print(f"Average Datetime: {pa.to_datetime(avg_datetime)}")
```

take the table and just take the mean of the date

##What was the average outside temperature on '09-11-2018'?

```python
df['noted_date'] = pa.to_datetime(df.noted_date, format='%m-%d-%Y')
date_to_filter = '09-11-2018'

filtered_df = df[(df['noted_date'].dt.date == pa.to_datetime(date_to_filter).date()) & (df['out/in'] == 'Out')]
average_outside_temp = filtered_df['temp'].mean()
print(f"Average outside temperature on {date_to_filter}: {average_outside_temp}")
```

Take the dataframe and format it to how the date is on the question. Select the date and make a new dataframe where its just the speciffied date and only the rows where its outside temp. Grab the mean from that and then print it.
