```python
import requests
import pandas as pa
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Chrome'}
r = requests.get('https://wiki.uooutlands.com/Creatures_List', headers=headers)
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

```python
tables = html_soup.find_all('table')
len(tables)
```

```python
from io import StringIO
import pandas as pa

df = pa.read_html(StringIO(str(tables)))[0]

df
```
