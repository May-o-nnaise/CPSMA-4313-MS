##1. Identify inccorect types

```python
data = {"age": ["20", "21", "19"], #should be an int
        "gpa": ["3.5", "3.7", "3.9"], #should be a float
        "enrolled": ["True", "False", "True"]} #should be a boolean
```

##2. convert to proper types

```python
data = {"age": [20, 21, 19],
        "gpa": [3.5, 3.7, 3.9],
        "enrolled": [True, False, True]}
#v alternatively, you could also do this v
data['age'] = [int(x) for x in data['age']]
data['gpa'] = [float(x) for x in data['gpa']]
data['enrolled'] = [bool(x) for x in data['enrolled']]
```

##3. Compute average GPA

```python
avgGPA = sum([float(x) for x in data['gpa']]) / 3
print(avgGPA)
```
