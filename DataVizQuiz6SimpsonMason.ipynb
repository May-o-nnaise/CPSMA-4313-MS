```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/refs/heads/main/Data_Sets/Activity_Dataset_V1.csv')
df
```

#Quiz 6: High Dimensional Data

```python
df['distance_normalized_alpha'] = df['distance'] / df['distance'].max()

ax = sns.scatterplot(data = df,
                x = 'calories',
                y = 'avg_heart_rate',
                hue = 'workout_type',
                style= 'workout_type',
                size = 'time')
if ax.collections:
    ax.collections[0].set_alpha(df['distance_normalized_alpha'].values)

sns.move_legend(ax, loc='upper left', bbox_to_anchor=(1, 1))
```
