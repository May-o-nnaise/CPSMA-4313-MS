##Import Data Set

```python

%%bigquery --project dataviz-488505
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

##Compute the average time for a trip based on starting point

```python

%%bigquery --project dataviz-488505
SELECT AVG(duration_minutes) AS avg_time, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = '10th/Red River'
GROUP BY start_station_name
```

I took the average of the minutes duration and the start station name where the station is the 10th Red River and i grouped it by start station name but I think that last part wasnt necessary but i had fun. This combination told me that the average trip from 10th Red River is 31.151388 minutes long.

##Compute how many trips start at each starting point

```python
%%bigquery --project dataviz-488505
SELECT
  COUNT(start_station_name) AS num, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```

To make this one, I copied over the Count one from the last quiz, removed the * and replaced it with start station name then edited it to be where it just groups by start station name. This is also kinda interesting because I didn't limit the data which means that there is only 201 stations that were recorded.

##Which starting point has the most trips

```python
%%bigquery --project dataviz-488505
SELECT
  COUNT(start_station_name) AS num, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY num DESC
LIMIT 1
```

To make this one I copied the last question and just added order by descending and limited it to one. Naming the count num helped me becaues now i could just type order by num desc and english wise that makes a lot of sense.

Also it looks like 21st speedway at PCL has the most trips at 108559.
