##2. Load Libraries

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

##3. Dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv')
df
```

##4. Examine Data

```python
dfg = df.groupby(['Class'])['Class'].count()
dfg
```

Looks like its an even spread of 50 for each type of Iris.

##5. Bar Chart

```python
dfg.plot(kind='bar', title='Types of Iris', ylabel='Number of Classes',
         xlabel='Class', figsize=(16, 5))
```

##6. Visualizations

```python
plt.figure(figsize=(16, 5))
plt.plot(df['SepalLength'], label='Sepal Length')
```

For my visualization I made a line graph of the sapial lengths but its not ordered in any unique way so its just starting from the first entry on the github data. So how about I order it?

```python
sorted_data = df.sort_values(by=['SepalLength'], ascending=True)
#print(sorted_data.head())
plt.plot(sorted_data['SepalLength'], label='Sepal Length')
```

Thats kinda interesting. It looks like this because there are some that have the same amount of entries with the same length so it is spread out as there are some spots where there is a really long line.
