```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv')
df
```

```python
plt.figure(figsize=(8, 6))
for name, group in df.groupby('Class'): #GENERATES COLOR
    plt.scatter(group['SepalLength'], group['SepalWidth'], label=name)
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.title('Scatter Plot of SepalLength vs SepalWidth by Class')
plt.legend()
plt.show()
```

In order to make the graph with color, you needed "for name, group in df.groupby('Class'):" and this adds the color because it uses a for to sort the data, then it makes a graph for each one and adds it to the one singular graph that is shown.
